```python
import pandas as pd

train_path = '/home/sk/Downloads/insurance_claim/train.csv'
test_path = '/home/sk/Downloads/insurance_claim/test.csv'

train_data = pd.read_csv(train_path) 
test_data = pd.read_csv(test_path) 

dummies_train_data = pd.get_dummies(train_data)
dummies_test_data = pd.get_dummies(test_data)
```

```python
dummies_train_data.head()
```

```python
features = dummies_train_data.columns.tolist()[1:]
```

```python
features.remove('target')
```

```python
features
```

```python
y = dummies_train_data.target
X = dummies_train_data[features]
```

```python
X
```

```python
from sklearn.model_selection import train_test_split

train_X, val_X, train_y, val_y = train_test_split(X, y, train_size=0.8, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()

model.fit(train_X, train_y)
```

```python
predictions = model.predict(val_X)
```

```python
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(val_y, predictions))
```

```python
model.fit(X, y)
```

```python
test_X = dummies_test_data[features]

test_predictions = model.predict(test_X)
```

```python
output_ap01 = pd.DataFrame({'id': test_data.id, 'target': test_predictions})
output_ap01.to_csv('/home/sk/Downloads/insurance_claim/my_submission_ap01.csv', index=False)
```
