```python
import pandas as pd

train_path = '/home/sk/Downloads/insurance_claim/train.csv'
test_path = '/home/sk/Downloads/insurance_claim/test.csv'

train_data = pd.read_csv(train_path) 
test_data = pd.read_csv(test_path) 
```

```python
# All categorical columns
object_cols = [col for col in train_data.columns if train_data[col].dtype == "object"]

from sklearn.preprocessing import OrdinalEncoder

ordinal_encoder = OrdinalEncoder()

train_data[object_cols] = ordinal_encoder.fit_transform(train_data[object_cols])

test_data[object_cols] = ordinal_encoder.transform(test_data[object_cols])
```

```python
features = train_data.columns.tolist()[1:-1]
```

```python
y = train_data.target
X = train_data[features]
```

```python
from sklearn.model_selection import train_test_split

train_X, val_X, train_y, val_y = train_test_split(X, y, train_size=0.8, test_size=0.2)
```

```python
from catboost import CatBoostRegressor
```

```python
model = CatBoostRegressor()
```

```python
model.fit(train_X, train_y)
```

```python
from sklearn.metrics import mean_absolute_error

preds_val = model.predict(val_X)
mae = mean_absolute_error(val_y, preds_val)
mae
```

```python
model.fit(X, y)
```

```python
test_X = test_data[features]

final_predict_ap05 = model.predict(test_X)
```

```python
output_ap05 = pd.DataFrame({'id': test_data.id, 'target': final_predict_ap05})
output_ap05.to_csv('/home/sk/Downloads/insurance_claim/my_submission_ap05.csv', index=False)
```
