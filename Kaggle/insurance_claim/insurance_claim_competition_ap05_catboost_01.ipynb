```python
import pandas as pd

train_path = '/home/sk/Downloads/insurance_claim/train.csv'
test_path = '/home/sk/Downloads/insurance_claim/test.csv'

train_data = pd.read_csv(train_path) 
test_data = pd.read_csv(test_path) 
```

```python
# All categorical columns
object_cols = [col for col in train_data.columns if train_data[col].dtype == "object"]

from sklearn.preprocessing import OrdinalEncoder

ordinal_encoder = OrdinalEncoder()

train_data[object_cols] = ordinal_encoder.fit_transform(train_data[object_cols])

test_data[object_cols] = ordinal_encoder.transform(test_data[object_cols])
```

```python
features = train_data.columns.tolist()[1:-1]
```

```python
y = train_data.target
X = train_data[features]
```

```python
X
```

```python
from sklearn.model_selection import train_test_split

train_X, val_X, train_y, val_y = train_test_split(X, y, test_size=0.3, shuffle=True)
```

```python
from catboost import CatBoostRegressor

model = CatBoostRegressor(iterations=2000,
                          task_type="GPU", 
                          devices='0:1', 
                          use_best_model=True)
```

```python
model.fit(train_X, train_y, eval_set = (val_X, val_y), early_stopping_rounds=100)
```

Check models MAE

```python
train_pred = model.predict(val_X)
```

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(val_y, train_pred)
mae
```

Evaluate model

Creating sumbission data

```python
test_X = test_data[features]

final_predict_ap05_01 = model.predict(test_X)
```

```python
output_ap05_01 = pd.DataFrame({'id': test_data.id, 'target': final_predict_ap05_01})
output_ap05_01.to_csv('/home/sk/Downloads/insurance_claim/my_submission_ap05_01.csv', index=False)
```
