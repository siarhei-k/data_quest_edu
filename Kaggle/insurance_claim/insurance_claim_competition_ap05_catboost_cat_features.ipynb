```python
import pandas as pd

train_path = '/home/sk/Downloads/insurance_claim/train.csv'
test_path = '/home/sk/Downloads/insurance_claim/test.csv'

train_data = pd.read_csv(train_path) 
test_data = pd.read_csv(test_path) 
```

```python
features = train_data.columns.tolist()[1:-1]
```

```python
categorical_features = features[0:10]
```

```python
y = train_data.target
X = train_data[features]
```

```python
from sklearn.model_selection import train_test_split

train_X, val_X, train_y, val_y = train_test_split(X, y, train_size=0.8, test_size=0.2)
```

```python
from catboost import CatBoostRegressor

model = CatBoostRegressor(iterations=3000, task_type="GPU", devices='0:1', cat_features = categorical_features)
```

```python
model.fit(train_X, train_y)
```

```python
train_pred = model.predict(val_X)
```

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(val_y, train_pred)
mae
```
