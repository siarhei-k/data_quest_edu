```python
import pandas as pd

train_path = '/home/sk/Downloads/insurance_claim/train.csv'
test_path = '/home/sk/Downloads/insurance_claim/test.csv'

train_data = pd.read_csv(train_path) 
test_data = pd.read_csv(test_path) 
```

```python
# All categorical columns
object_cols = [col for col in train_data.columns if train_data[col].dtype == "object"]

from sklearn.preprocessing import OrdinalEncoder

ordinal_encoder = OrdinalEncoder()

train_data[object_cols] = ordinal_encoder.fit_transform(train_data[object_cols])

test_data[object_cols] = ordinal_encoder.transform(test_data[object_cols])
```

```python
features = train_data.columns.tolist()[1:-1]
```

```python
y = train_data.target
X = train_data[features] / 255
```

```python
# # Scale to [0, 1]
# max_ = X.max(axis=0)
# min_ = X.min(axis=0)
# X = (X - min_) / (max_ - min_)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split

train_X, val_X, train_y, val_y = train_test_split(X, y, test_size=0.3, shuffle=True)
```

```python
print(train_X.shape)
```

```python
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential([
    layers.Dense(1024, activation='relu', input_shape=[24]),
    layers.Dense(768, activation='relu'),
    layers.Dense(512, activation='relu'),
    layers.Dense(256, activation='relu'),
    layers.Dense(1),
])
```

```python
model.compile(
    optimizer='Adam',
    loss='mae',
)
```

```python
model.fit(
    train_X, train_y,
    validation_data=(val_X, val_y),
    batch_size=210,
    epochs=30,
)
```

```python
train_pred = model.predict(val_X)
```

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(val_y, train_pred)
mae
```
