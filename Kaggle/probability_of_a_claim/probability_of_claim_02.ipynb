```python
import pandas as pd

train_path = '/home/sk/Downloads/tabular-playground-series-sep-2021/train.csv'
test_path = '/home/sk/Downloads/tabular-playground-series-sep-2021/test.csv'

train_data = pd.read_csv(train_path) 
test_data = pd.read_csv(test_path)
```

```python
features = train_data.columns.tolist()[1:-1]
```

```python
train_data.dtypes.value_counts()
```

```python
for column in train_data.columns.tolist()[1:]:
    train_data[column].fillna(train_data[column].median())

    
for column in test_data.columns.tolist()[1:]:
    test_data[column].fillna(test_data[column].median())
```

```python
test_data
```

```python
X = train_data[features]
y = train_data.claim
test_X = test_data[features]
```

```python
X
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaled_features = scaler.fit_transform(X.values)


X = pd.DataFrame(scaled_features, index=X.index, columns=X.columns)
```

```python
scaled_features_test = scaler.transform(test_X.values)

test_X = pd.DataFrame(scaled_features_test, index=test_X.index, columns=test_X.columns)
```

```python
X
```

```python
test_X 
```

```python
from sklearn.model_selection import train_test_split

train_X, val_X, train_y, val_y = train_test_split(X, y, test_size=0.2, shuffle=True)
```

```python
from catboost import CatBoostRegressor

model = CatBoostRegressor(iterations=2000,
                          task_type="GPU", 
                          devices='0:1', 
                          use_best_model=True)
```

```python
model.fit(train_X, train_y, eval_set = (val_X, val_y), early_stopping_rounds=100)
```

```python
train_pred = model.predict(val_X)
```

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(val_y, train_pred)
mae
```

```python
final_predict = model.predict(test_X)
```

```python
output = pd.DataFrame({'id': test_data.id, 'claim': final_predict})
output.to_csv('/home/sk/Downloads/tabular-playground-series-sep-2021/my_submission_02.csv', index=False)
```
