```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df = pd.read_csv(r'C:\Users\Siarhei\Downloads\states (1).csv')
X = df['hs_grad']
y = df['poverty']

X_mean = X.mean()
y_mean = y.mean()

N = X.shape[0] - 1

SDx = np.sqrt(sum([(i-X_mean)**2 for i in X])/N)
SDy = np.sqrt(sum([(i-y_mean)**2 for i in y])/N)

cov = sum([(i-X_mean)*(j-y_mean) for i, j in zip(X, y)])/N

Rxy = cov/(SDx*SDy)

b1 = (SDy/SDx) * Rxy
b0 = y_mean - X_mean*b1

x_pl = np.linspace(78, 92, 10)  
y_pl = b0 + b1*x_pl

plt.scatter(X, y)
plt.plot(x_pl, y_pl, color="blue")
plt.show()

print('b1: ', b1)
print('b0: ', b0)
print('cov: ', cov)
print('Rx: ', Rxy)
print('Sdx: ', SDx)
print('SDy: ', SDy)
print('X_mean: ', X_mean)
print('y_mean: ', y_mean)
print('N: ', X.shape[0])
```
